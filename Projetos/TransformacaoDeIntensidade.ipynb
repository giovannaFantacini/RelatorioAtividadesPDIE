##### Aplicar a transformação logarítmica, testar vários valores para o parâmetro c "s = c log (1 + r)"

```python
def transform_Log(image, c):
    log_image = c * (np.log(image + 1)) 
    log_image = np.array(log_image, dtype = np.uint8) 
    return log_image
   

fig = plt.figure(figsize=(10, 5))
plt1 = plt.subplot(1, 3, 1)
plt2 = plt.subplot(1, 3, 2)
plt3 = plt.subplot(1, 3, 3)
plt1.title.set_text("Original")
plt2.title.set_text("Transformação Logaritmica, c = 5")
plt3.title.set_text("Transformação Logaritmica, c = 1")
plt1.imshow(image1, cmap='gray')
log_image05 = np.array(transform_Log(image1, 5), dtype = np.uint8)
plt2.imshow(log_image05, cmap='gray')
log_image1 = np.array(transform_Log(image1, 1), dtype = np.uint8)
plt3.imshow(log_image1,  cmap='gray')
plt.subplots_adjust(wspace=0.5)

fig = plt.figure(figsize=(10, 5))
plt1 = plt.subplot(1, 3, 1)
plt2 = plt.subplot(1, 3, 2)
plt3 = plt.subplot(1, 3, 3)
plt1.title.set_text("Original")
plt2.title.set_text("Transformação Logaritmica, c = 5")
plt3.title.set_text("Transformação Logaritmica, c = 1")
plt1.imshow(image2,  cmap='gray')
log_image05 = np.array(transform_Log(image2, 5), dtype = np.uint8)
plt2.imshow(log_image05,  cmap='gray')
log_image1 = np.array(transform_Log(image2, 1), dtype = np.uint8)
plt3.imshow(log_image1,  cmap='gray')
plt.subplots_adjust(wspace=0.5)

```

##### Implemente a representação de cada plano de bits das imagens


```python
def bits_planning(image, sigBit):
    width, height = image.shape
    binaryImage= np.zeros((width, height))
    
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
           bit = str(bin(image[i][j]))
           bit = bit[2:]
           bit = bit.rjust(8,'0')
           binaryImage[i][j] = int(bit[sigBit])
    return binaryImage

fig = plt.figure(figsize=(15, 10))
plt1 = plt.subplot(2, 4, 1)
plt2 = plt.subplot(2, 4, 2)
plt3 = plt.subplot(2, 4, 3)
plt4 = plt.subplot(2, 4, 4)
plt5 = plt.subplot(2, 4, 5)
plt6 = plt.subplot(2, 4, 6)
plt7 = plt.subplot(2, 4, 7)
plt8 = plt.subplot(2, 4, 8)
plt1.title.set_text("Significant Bit = 8")
plt2.title.set_text("Significant Bit = 7")
plt3.title.set_text("Significant Bit = 6")
plt4.title.set_text("Significant Bit = 5")
plt5.title.set_text("Significant Bit = 4")
plt6.title.set_text("Significant Bit = 3")
plt7.title.set_text("Significant Bit = 2")
plt8.title.set_text("Significant Bit = 1")
plt1.imshow(bits_planning(image1, 0), cmap='gray')
plt2.imshow(bits_planning(image1, 1), cmap='gray')
plt3.imshow(bits_planning(image1, 2), cmap='gray')
plt4.imshow(bits_planning(image1, 3), cmap='gray')
plt5.imshow(bits_planning(image1, 4), cmap='gray')
plt6.imshow(bits_planning(image1, 5), cmap='gray')
plt7.imshow(bits_planning(image1, 6), cmap='gray')
plt8.imshow(bits_planning(image1, 7), cmap='gray')
plt.show()


fig = plt.figure(figsize=(15, 10))
plt1 = plt.subplot(2, 4, 1)
plt2 = plt.subplot(2, 4, 2)
plt3 = plt.subplot(2, 4, 3)
plt4 = plt.subplot(2, 4, 4)
plt5 = plt.subplot(2, 4, 5)
plt6 = plt.subplot(2, 4, 6)
plt7 = plt.subplot(2, 4, 7)
plt8 = plt.subplot(2, 4, 8)
plt1.title.set_text("Significant Bit = 8")
plt2.title.set_text("Significant Bit = 7")
plt3.title.set_text("Significant Bit = 6")
plt4.title.set_text("Significant Bit = 5")
plt5.title.set_text("Significant Bit = 4")
plt6.title.set_text("Significant Bit = 3")
plt7.title.set_text("Significant Bit = 2")
plt8.title.set_text("Significant Bit = 1")
plt1.imshow(bits_planning(image2, 0), cmap='gray')
plt2.imshow(bits_planning(image2, 1), cmap='gray')
plt3.imshow(bits_planning(image2, 2), cmap='gray')
plt4.imshow(bits_planning(image2, 3), cmap='gray')
plt5.imshow(bits_planning(image2, 4), cmap='gray')
plt6.imshow(bits_planning(image2, 5), cmap='gray')
plt7.imshow(bits_planning(image2, 6), cmap='gray')
plt8.imshow(bits_planning(image2, 7), cmap='gray')
plt.show()

```

##### Implementar a equalização do histograma  

```python
def instantiate_histogram():    
    hist_array= []
    
    for i in range(0,256):
        hist_array.append(str(i))
        hist_array.append(0)
    
    hist_dct = {hist_array[i]: hist_array[i + 1] for i in range(0, len(hist_array), 2)} 
    
    return hist_dct

def count_intensity_values(hist, img):
    for row in img:
        for column in row:
            hist[str(int(column))] = hist[str(int(column))] + 1
     
    return hist

def plot_hist(hist, hist2=''):
    if hist2 != '':
        figure, axarr = plt.subplots(1,2, constrained_layout = True)
        figure.suptitle("Histogram x Equalized Histogram", fontsize=16)
        axarr[0].bar(hist.keys(), hist.values())
        axarr[1].bar(hist2.keys(), hist2.values())
    else:
        plt.bar(hist.keys(), hist.values())
        plt.xlabel("Níveis intensidade")
        ax = plt.gca()
        ax.axes.xaxis.set_ticks([])
        plt.grid(True)
        plt.show()


def get_hist_proba(hist, n_pixels):
    hist_proba = {}
    for i in range(0, 256):
        hist_proba[str(i)] = hist[str(i)] / n_pixels
    
    return hist_proba

def get_accumulated_proba(hist_proba): 
    acc_proba = {}
    sum_proba = 0
    
    for i in range(0, 256):
        if i == 0:
            pass
        else: 
            sum_proba += hist_proba[str(i - 1)]
            
        acc_proba[str(i)] = hist_proba[str(i)] + sum_proba
    return acc_proba

def get_new_gray_value(acc_proba):
    new_gray_value = {}
    
    for i in range(0, 256):
        new_gray_value[str(i)] = np.ceil(acc_proba[str(i)] * 255)
    return new_gray_value

def equalize_hist(img, new_gray_value):
    for row in range(img.shape[0]):
        for column in range(img.shape[1]):
            img[row][column] = new_gray_value[str(int(img[row] [column]))]
            
    return img

histogram = instantiate_histogram()
    
histogram = count_intensity_values(histogram, image1)

n_pixels = image1.shape[0] * image1.shape[1]
hist_proba = get_hist_proba(histogram, n_pixels)
accumulated_proba = get_accumulated_proba(hist_proba)
new_gray_value = get_new_gray_value(accumulated_proba)
eq_img = equalize_hist(image1.copy(), new_gray_value)

eq_histogram = instantiate_histogram()

eq_histogram = count_intensity_values(eq_histogram, eq_img)

plot_hist(histogram, hist2=eq_histogram)
plt.show()

fig = plt.figure(figsize=(10, 5))
plt1 = plt.subplot(1, 2, 1)
plt2 = plt.subplot(1, 2, 2)
plt1.title.set_text("Original")
plt2.title.set_text("Equalizada")
plt1.imshow(image1, cmap='gray')
plt2.imshow(eq_img, cmap='gray')
plt.subplots_adjust(wspace=0.5)
```
